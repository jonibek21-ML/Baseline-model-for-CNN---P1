# 📓 CIFAR-10 Image Classification using CNN (PyTorch)

Bu notebookda PyTorch yordamida CIFAR-10 dataset ustida CNN model train qilinadi.

Pipeline:

1. Dataset yuklash
2. Data augmentation
3. DataLoader yaratish
4. CNN model qurish
5. Training loop yozish
6. Loss monitoring

## 📦 1️⃣. Kerakli kutubxonalarni import qilish

Bu qismda model qurish va dataset bilan ishlash uchun kerakli kutubxonalar yuklanadi:

- torch → deep learning framework
- torchvision → dataset + transforms
- nn → neural network layerlar
- transforms → data preprocessing

````python
import numpy
import torch
import torch.nn as nn
import torchvision.transforms
import torchvision.transforms as transforms
````

## 🔄 2️⃣. Training dataset uchun Data Augmentation

Training datasetga augmentation qo‘llaymiz:

Sababi:

- model generalization oshadi
- overfitting kamayadi
- dataset ko‘paygandek effekt beradi

Ishlatilayotgan augmentationlar:

- RandomRotation
- RandomCrop
- RandomHorizontalFlip
- Tensor conversion

Test datasetga esa transform section yetarli:

````python
transform_tr = transforms.Compose([
    transforms.RandomCrop(32,padding=4),
    transforms.RandomRotation(10),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        (0.4914, 0.4822, 0.4465),
        (0.2470, 0.2435, 0.2616)
    )
])
transform_te = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        (0.4914, 0.4822, 0.4465),
        (0.2470, 0.2435, 0.2616)
    )
])
````

## 📊 3️⃣. CIFAR-10 datasetni yuklash

Bu qismda CIFAR-10 dataset train va test qismlarga ajratilib yuklanadi.

Dataset xususiyatlari:

- 10 ta class
- 60 000 ta rasm
- rasm o‘lchami: 32×32
- RGB format (3 channel)

````python
data_tr = torchvision.datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transform_tr)

data_te = torchvision.datasets.CIFAR10(
    root='./data',
    train=False,
    download=True,
    transform=transform_te)
````

## 🧪 4️⃣. DataLoader yaratish

DataLoader modelni mini-batch training qilish imkonini beradi.

Batch size:

128

Shuffle:

Training dataset uchun True
Test dataset uchun False

````python
load_tr = torch.utils.data.DataLoader(
    data_tr,
    batch_size=128,
    shuffle=True)

load_te = torch.utils.data.DataLoader(
    data_te,
    batch_size=128,
    shuffle=False)
````

## 🧠 5️⃣. CNN model arxitekturasi

Model quyidagi layerlardan tashkil topgan:

Conv2D → feature extraction

BatchNorm → training stabilizatsiya qiladi

ReLU → nonlinear activation qo‘shadi

MaxPool → feature map o‘lchamini kamaytiradi

Dropout → overfittingni kamaytiradi

Linear → classification layer

````python
model = nn.Sequential(
    nn.Conv2d(3, 32, 3, padding=1),
    nn.BatchNorm2d(32),
    nn.ReLU(),
    nn.Conv2d(32, 32, 3, padding=1),
    nn.BatchNorm2d(32),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.25),
    
    nn.Conv2d(32, 64, 3, padding=1),
    nn.BatchNorm2d(64),
    nn.ReLU(),
    nn.Conv2d(64, 64, 3, padding=1),
    nn.BatchNorm2d(64),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.25),

    nn.Conv2d(64, 128, 3, padding=1),
    nn.BatchNorm2d(128),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Dropout(0.25),
    
    nn.Flatten(),
    nn.Linear(128*4*4,512),
    nn.ReLU(),
    nn.Dropout(0.3),
        
    nn.Linear(512,10)
)
````

## 🚀 6️⃣. Device, Loss Function va Optimizer sozlash

Bu qismda model training uchun kerakli uchta asosiy komponent sozlanadi:

- Device (CPU yoki GPU)
- Loss function
- Optimizer

### Device tanlash

Quyidagi kod modelni avtomatik ravishda GPU mavjud bo‘lsa GPU’da, aks holda CPU’da ishlashini ta’minlaydi:

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
optm = torch.optim.Adam(model.parameters(),lr=0.001)
scheduler = torch.optim.lr_scheduler.StepLR(
    optm,
    step_size=10,
    gamma=0.5
)
crit = nn.CrossEntropyLoss()
````

## 🔁 7️⃣. Training loop

Training loop quyidagi bosqichlardan iborat:

1. Forward pass
2. Loss calculation
3. Backward pass
4. Weight update

Har epoch oxirida loss qiymati chiqariladi.

````python
for epoch in range(20):
    model.train()
    running_loss = 0
    for images, labels in load_tr:
        images = images.to(device)
        labels = labels.to(device)
        optm.zero_grad()
        outputs = model(images)
        loss = crit(outputs, labels)
        loss.backward()
        optm.step()
        running_loss += loss.item()
    scheduler.step()
    print(f"Epoch {epoch+1} Loss: {running_loss:.3f}")
````

## 📈 8️⃣.Modelni Test Datasetda Baholash (Evaluation Phase)
Bu qismda model test dataset ustida baholanadi va umumiy **accuracy** hisoblanadi.
Modelni baholaydigan boshqa metricalarni ham ishlatsa bo'ladi >>

````python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for image,label in load_te:
        image = image.to(device)
        label = label.to(device)
        pr = model(image)
        _, predicted = torch.max(pr, 1)
        total += label.size(0)
        correct += (predicted == label).sum().item()
print(f"Accuracy:{(correct / total) * 100}")
````
